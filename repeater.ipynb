```python
buttonDictionary = {
    "Button.unknown":Button.unknown,
    "Button.left":Button.left,
    "Button.Middle":Button.middle,
    "Button.right":Button.right}
```

```python
log_dir = open(r"keylog.txt","w+")
full=[]
def on_click(x,y,button,pressed):
	if pressed:
		time_shot_press=int(time.time())
		btn=str(button)
		item=[time_shot_press,x,y,btn]
		full.append(item)
mouse_listener = Listener1(on_click=on_click)
mouse_listener.start()
def on_press(key):
    return True
def on_release(key):
	time_shot=int(time.time())
	if type(key)==Key:
		key_shot=str(key)
	else:
		key_shot=key.char	
	item=[time_shot,key_shot]
	full.append(item)
	if key == Key.esc:
		mouse_listener.stop()
		full_instructions=full[:-1]
		time_list=[0]
		for i in range(len(full_instructions)-1):
			if full_instructions[i + 1][0] - full_instructions[i][0]>3:
				time_list.append(full_instructions[i + 1][0] - full_instructions[i][0])
			elif len(full_instructions[i])==4:
				time_list.append(2)
			else:
				time_list.append(0.25)
		for i in range(len(full_instructions)):
			full_instructions[i][0]=time_list[i]
		log_dir.write(str(full_instructions))
		log_dir.close()
		return False

with Listener(on_press=on_press,on_release=on_release) as listener:
    listener.join()
```

```python
import ctypes
PROCESS_PER_MONITOR_DPI_AWARE = 2
ctypes.windll.shcore.SetProcessDpiAwareness(PROCESS_PER_MONITOR_DPI_AWARE)
k=Controller()
mouse = Controller1()
n=int(input("Enter the number of times you want to repeat"))
file=open(r"keylog.txt")
res=""
for i in file:
	res+=i
	res = ast.literal_eval(res)
for j in range(n):
	for i in res:
		time.sleep(i[0])
		if len(i)==4:
			mouse.position=(i[1],i[2])
			mouse.click(buttonDictionary[i[3]],1)
		else:
			if i[1].startswith("Key"):
				k.press(keyDictionary[i[1]])
				k.release(keyDictionary[i[1]])
			else:
				k.type(i[1].lstrip())
```
